Let's assume we take a sample of 50 people, record their height, and then repeat
this process 30 times. We can then plot the averages of each sample and observe
the distribution.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
average_height = []
for i in range(30):
    sample50 = np.random.normal(183, 10, 50)
    average_height.append(sample50.mean())
plt.hist(average_height,20)
plt.show()
```

Z-test vs T-**test**

```python
from scipy import stats
c1 = np.array([75.0, 84.0, 53.0, 70.0])
c2 = np.array([75.0, 85.0, 53.0, 70.0])
stats.ttest_ind(c1,c2)
```

Studying the **Titanic**

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/Questions/titanic.csv')
print("checking each data set having null value or not \n ")
df['Pclass'].isnull().value_counts()
df['Survived'].isnull().value_counts()
print("Passengers survived in each class \n")
survivors = df.groupby('Pclass')['Survived'].agg(sum)
print(survivors)
print("Total Passengers in each class \n")
total_passengers = df.groupby('Pclass')['PassengerId'].count()
survivor_percentage = survivors / total_passengers
print(survivor_percentage)
```

Box **plots**
A box plot is a very good plot to understand the spread, median, and outliers of data:

* Q3: This is the 75th percentile value of the data. It's also called the upper hinge.
* Q1: This is the 25th percentile value of the data. It's also called the lower hinge.
* Box: This is also called a step. It's the difference between the upper hinge and
the lower hinge
* Median: This is the midpoint of the data.
* Max: This is the upper inner fence. It is 1.5 times the step above Q3.
* Min: This is the lower inner fence. It is 1.5 times the step below Q1.
